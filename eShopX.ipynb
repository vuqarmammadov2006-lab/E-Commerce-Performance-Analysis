```python
# Loading the dataset using pandas

import pandas as pd
df = pd.read_csv('Pakistan Largest Ecommerce Dataset.csv',low_memory=False)
```

```python
# To see the Missing Values
df_missing = df.copy()

missing_data = df_missing.isnull().sum().sort_values(ascending=False).reset_index()
missing_data.columns = ['Columns Name', 'Missing Values']
missing_data['Percentage Missing Values'] = (missing_data['Missing Values'] / len(df_missing)) * 100
missing_data['Percentage Missing Values'] = missing_data['Percentage Missing Values'].apply(lambda x: f"{x:.0f}%")
missing_data
```

```python
#delete the unnamed collumns that do not have values inside 
#delete sales_commission code column because more than half of the column is missing and it is not very relevant for our analysis
#delete the m-y (month and the year of the transaction), because we already have sepereate columns for months and years
df = df.drop(
    ['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'sales_commission_code', 'M-Y'],
    axis=1,
    errors='ignore'
)
```

```python
#drop the rows where all of the values in columns are missing
#review the null values left
df = df.dropna(how='all')
df.isna().sum()
```

```python
# Because sku(stock keeping unit) and customer id are the primary keys(unique) of the table we are going to delete the rows with these values missing
df.dropna(subset=['sku', 'Customer ID'], inplace=True)

# for status, category_name_1, we are going to fill with it with the mode, because they are categories 
# for Customer Since, because it is a date column, it is logical to take the last known valid date
df['status'] = df['status'].fillna(df['status'].mode()[0])
df['category_name_1'] = df['category_name_1'].fillna(df['category_name_1'].mode()[0])
df['Customer Since'] = df['Customer Since'].bfill()
```

```python
# Renaming columns for better readability and documentation
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#Rename certain columns for better readibility
df = df.rename(columns={
    '_mv_': 'market_value',
    'fy': 'fiscal_year',
    'sku': 'stock_keeping_unit', 
    'category_name_1': 'category_name',
})
```

```python
#Based on the head of the data, it is logical to assume that the created_at and working_date might be same for each transaction 
(df['working_date'] == df['created_at']).all()
```

```python
#rewrite the grand_total amount formula, to make sure it is correct for each row, since it is a calculation of existing rows
df["grand_total"] = df['price'] * df['qty_ordered'] - df['discount_amount']
df.head(10)
```

```python
#For simplicity, we will leave just one column called the transaction date 
df = df.rename(columns={
    'working_date': 'transaction_date'})
df = df.drop(['created_at'],axis=1,errors='ignore')
df.columns
```

```python
# We are going to start reviewing categorical columns
df['status'].value_counts()
```

```python
# In our case complete and received is the same status, just recorded differently, so is order_refunded and refund, and pedning and pending_paypal
# cod (cash on delivery) is a payment_method not an order_status, we are going to categoize them as complete

# Replace invalid / missing values with the mode - which is complete, because the number is not significant
df['status'] = df['status'].replace({'\\N': 'complete'})

# Merge pending-type statuses
pending_statuses = ['pending_paypal', 'payment_review', 'holded']
df.loc[df['status'].isin(pending_statuses), 'status'] = 'Pending'

#Merge complete-type statuses
completed_statuses = ['complete', 'cod', 'received']
df.loc[df['status'].isin(completed_statuses), 'status'] = 'Completed'

# Merge refunded statuses
refunded_statuses = ['refund', 'order_refunded']
df.loc[df['status'].isin(refunded_statuses), 'status'] = 'Refunded'

# Normalize capitalization for consistency
df['status'] = df['status'].str.capitalize()

# Check the cleaned value counts
print(df['status'].value_counts(dropna=False))
```

```python
#Since there a lot of \N values we are going to replace them with 'Unknown'
df['category_name'] = df['category_name'].replace({'\\N': 'Unknown'})
```

```python
df['bi_status'].value_counts()
```

```python
df['bi_status'] = df['bi_status'].replace({'#REF!': 'Net'}) #replace with the mode - which is 'Net'
df['bi_status'].value_counts()
```

```python
#Convert the types of columns
df.dtypes
```

```python
# Convert into Datetime Datatype
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce')
df['customer_since'] = pd.to_datetime(df['customer_since'], errors='coerce')
# Convert specified columns into categorical type
categorical_columns = ['status', 'payment_method', 'category_name', 'bi_status', 'stock_keeping_unit']
df[categorical_columns] = df[categorical_columns].astype('category')
# Numeric conversion and handle missing values if appears during errors='coerce'

# Convert 'increment_id' to numeric and coerce invalid entries to NaN
# (Invalid entries, like strings, are replaced with NaN using errors='coerce')
df['increment_id'] = pd.to_numeric(df['increment_id'], errors='coerce')
# Drop rows where 'increment_id' is NaN
df.dropna(subset=['increment_id'], inplace=True)
# Convert the column to integer after dropping NaN
df['increment_id'] = df['increment_id'].astype(int)

# Convert price column to float
df['price']= pd.to_numeric(df['price'], errors='coerce')
df['price']= df['price'].fillna(df['price'].mean()).astype(float)

# Convert qty_ordered column to int
df['qty_ordered'] = pd.to_numeric(df['qty_ordered'], errors='coerce')
df['qty_ordered']= df['qty_ordered'].fillna(df['qty_ordered'].median()).astype(int)

# Convert customer_id column to int
df['customer_id'] = pd.to_numeric(df['customer_id'], errors='coerce')
df.dropna(subset=['customer_id'], inplace=True)
df['customer_id'] = df['customer_id'].astype(int)

# Convert grand_total column to int
df['grand_total']= pd.to_numeric(df['grand_total'], errors='coerce')
df['grand_total']= df['grand_total'].fillna(df['grand_total'].median()).astype(float)

# Convert item_id column to int
df['item_id'] = pd.to_numeric(df['item_id'], errors='coerce')
df.dropna(subset=['item_id'], inplace=True)
df['item_id'] = df['item_id'].astype(int)

# Convert market_value column to float
df['market_value'] = pd.to_numeric(df['market_value'], errors='coerce')
df['market_value']= df['market_value'].fillna(df['market_value'].mean()).astype(float)

# Convert discount_amount column to float
df['discount_amount'] = pd.to_numeric(df['discount_amount'], errors='coerce')
df['discount_amount']= df['discount_amount'].fillna(df['discount_amount'].median()).astype(float)

# Convert year column to int
df['year']= df['year'].astype(int)

# Convert month column to int
df['month']= df['month'].astype(int)
```
